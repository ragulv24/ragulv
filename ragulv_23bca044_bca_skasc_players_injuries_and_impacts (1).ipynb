```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)  # Changed 'data' to 'df'
plt.title('Histogram of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,  # Changed 'data' to 'df'
    x='fifa_rating',  # Replace with your desired column
    bins=15,
    kde=True,  # Add a Kernel Density Estimate curve
    color='coral',
    alpha=0.8
)
plt.title('Histogram of FIFA Rating with KDE', fontsize=16)
plt.xlabel('FIFA Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE plot using Seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,  # Changed 'data' to 'df' to use the DataFrame you created
    x='fifa_rating',  # Replace with your desired column
    shade=True,  # Adds shading under the curve
    color='blue',
    alpha=0.8,
    bw_adjust=0.5  # Bandwidth adjustment for smoothing
)
plt.title('KDE Plot of FIFA Rating', fontsize=16)
plt.xlabel('FIFA Rating', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Optional: KDE for multiple groups (e.g., 'position')
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,  # Changed 'data' to 'df' here as well
    x='age',  # Replace with your desired column
    hue='position',  # Group by a categorical column
    fill=True,  # Adds shaded areas for each group
    palette='viridis',
    alpha=0.6,
    common_norm=False  # Prevents normalizing across groups
)
plt.title('KDE Plot of Age by Position', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(title='Position', fontsize=10, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Example column for Bar Plot
bar_column = 'Team Name'

# Count the occurrences of each unique value in the bar_column
# Changed 'data' to 'df' to use the correct DataFrame
data_bar = df[bar_column].value_counts()

# Plot Bar Plot
plt.figure(figsize=(10, 6))
data_bar.plot(kind='bar', color='orange', alpha=0.8)
plt.title(f'Bar Plot for {bar_column}', fontsize=14)
plt.xlabel(bar_column, fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
box_column = 'FIFA rating'

# Drop rows with missing or invalid data for the selected column
# Changed 'data' to 'df' to use the DataFrame that holds the data
data_box = df[box_column].dropna()

# Plot Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=data_box, color='green', width=0.5)
plt.title(f'Box Plot for {box_column}', fontsize=14)
plt.xlabel(box_column, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Example columns for Strip Plot
strip_x_column = 'Position'
strip_y_column = 'FIFA rating'

# Drop rows with missing or invalid data for the selected columns
# Changed 'data' to 'df' to use the DataFrame that holds the data
data_strip = df.dropna(subset=[strip_x_column, strip_y_column]) # Changed 'data' to 'df'

# Plot Strip Plot
plt.figure(figsize=(10, 6))
sns.stripplot(x=strip_x_column, y=strip_y_column, data=data_strip, jitter=True, alpha=0.7, palette='viridis')
plt.title(f'Strip Plot of {strip_y_column} by {strip_x_column}', fontsize=14)
plt.xlabel(strip_x_column, fontsize=12)
plt.ylabel(strip_y_column, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Example columns for Join Plot
join_x_column = 'Age'
join_y_column = 'FIFA rating'

# Drop rows with missing or invalid data for the selected columns
# Changed 'data' to 'df' to use the DataFrame you created
data_join = df.dropna(subset=[join_x_column, join_y_column])

# Plot Join Plot
sns.jointplot(x=join_x_column, y=join_y_column, data=data_join, kind='scatter', color='purple', height=8)
plt.suptitle(f'Join Plot of {join_y_column} vs {join_x_column}', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Example column for Distribution Bars
dist_column = 'Age'

# Drop rows with missing or invalid data for the selected column
data_dist = df[dist_column].dropna() # Changed 'data' to 'df'

# Plot Distribution Bars
plt.figure(figsize=(10, 6))
sns.histplot(data_dist, bins=20, kde=False, color='skyblue', alpha=0.8)
plt.title(f'Distribution Bars for {dist_column}', fontsize=14)
plt.xlabel(dist_column, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Example columns for Relationship Plot
rel_x_column = 'Age'
rel_y_column = 'FIFA rating'

# Drop rows with missing or invalid data for the selected columns
# Changed 'data' to 'df' to use the DataFrame you created
data_rel = df.dropna(subset=[rel_x_column, rel_y_column]) # This line was changed, replacing 'data' with 'df'

# Plot Relationship Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=rel_x_column, y=rel_y_column, data=data_rel, hue='Position', alpha=0.7, palette='viridis')
plt.title(f'Relationship Plot of {rel_y_column} vs {rel_x_column}', fontsize=14)
plt.xlabel(rel_x_column, fontsize=12)
plt.ylabel(rel_y_column, fontsize=12)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Example columns for Facet Grid
facet_x_column = 'Age'
facet_y_column = 'FIFA rating'
facet_col_column = 'Position'

# Drop rows with missing or invalid data for the selected columns
# Changed 'data' to 'df' to use the DataFrame you created
data_facet = df.dropna(subset=[facet_x_column, facet_y_column, facet_col_column]) # Changed 'data' to 'df'

# Create Facet Grid
facet = sns.FacetGrid(data_facet, col=facet_col_column, col_wrap=4, height=4, sharex=True, sharey=True)
facet.map(sns.scatterplot, facet_x_column, facet_y_column, alpha=0.7)
facet.set_titles(col_template="{col_name}", row_template="{row_name}")
facet.set_axis_labels(facet_x_column, facet_y_column)
facet.tight_layout()
plt.show()
```
